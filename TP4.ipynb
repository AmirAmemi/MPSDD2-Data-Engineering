```python
import pandas as pd
import matplotlib.pyplot as plt
# Nouveau chemin du fichier de logs
log_file_path = 'logfile(4).log'
# Charger le fichier de logs
columns = ['ip_address', 'User', 'Date', 'Request', 'Status']
df = pd.read_csv(log_file_path, sep=' ', names=columns, parse_dates=['Date'], dayfirst=True)
# Display the first few rows of the DataFrame
print (df.head())
# Compter le nombre d'occurrences de chaque adresse IP
ip_counts = df['ip_address'].value_counts ()
print(ip_counts)
# Définir un seuil pour le nombre d'occurrences considéré comme suspect
seuil_suspect = 2
# Filtrer les adresses IP suspectes
adresses_ip_suspectes = ip_counts[ip_counts > seuil_suspect].index
```

```python
# Afficher les adresses IP suspectes
print ("Adresses IP suspectes :")
print (adresses_ip_suspectes)
# Tracer un diagramme pour visualiser les adresses IP
plt.bar(ip_counts.index, ip_counts.values)
plt.xlabel ('Adresse IP')
plt. ylabel ('Nombre d\'occurrences')
plt. title ('Occurrence des adresses IP dans les logs')
plt. show()
```
